```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
combine_data_df = pd.merge(city_data_df, ride_data_df, on='city')

# Display the data table for preview
combine_data_df.head()
combine_data_df
```

## Bubble Plot of Ride Sharing Data

```python
#Average Fare ($) Per City
grouped_city_fare_df = combine_data_df.groupby(['type', 'city']).mean()
```

```python
df1 = pd.DataFrame(data=grouped_city_fare_df).reset_index()
df1 = df1[['type', 'city', 'fare']]
df1
```

```python
#Total Number of Rides Per City
grouped_city_rides_df = combine_data_df.groupby(['type', 'city']).count()
```

```python
df2 = pd.DataFrame(data=grouped_city_rides_df).reset_index()
df2 = df2[['type', 'city', 'ride_id']]
df2
```

```python
#Total Number of Drivers Per City
grouped_city_drivers_df = combine_data_df.groupby(['type', 'city', 'driver_count']).sum()
```

```python
df3 = pd.DataFrame(data=grouped_city_drivers_df).reset_index()
df3 = df3[['type', 'city', 'driver_count']]
df3
```

```python
# add grids
plt.title('Pyber Ride Sharing Data (2016)')
plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare ($)')

plt.grid(True)
#plt.legend(loc='upper right')

# setting limit
#plt.xlim(45, 40)

# Incorporate a text label regarding circle size
plt.text(0.1, 0.5, 'Note:\n Circle size correlates with driver count per city.', ha='right', fontsize=10)

# Obtain the x and y coordinates for each of the three city types
plt.scatter(df1.fare, df2.ride_id, s=df3.driver_count)
```
